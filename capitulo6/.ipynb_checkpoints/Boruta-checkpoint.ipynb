# Nesse capítulo vou explorar o método de seleção de variáveis chamado Boruta. 
### Esse método se baseia em shadow features e nas features originais. Shadow features são cópias das variáveis originais, só que elas são embaralhadas aleatoriamente. Então, treinando um modelo que reproduz algum tipo de feature importance, como a Random Forest e a partir daí verificamos todas as features importantes, tanto das features originais tanto das shadow features. 

### Aqui, verificamos qual é a maior feature importance das shadow feature e guardamos esse valor. Se alguma feature importance das features originais for menor do que esse valor, então essa feature perde não ganha nenhum ponto e se for maior, ganha um ponto. Esse processo é realizado em várias iterações e no final registramos as pontuações de todas as features originais. Em seguinda, calculamos o valor-p para cada feature original, se a quantidade de ponto for muito baixa ou nula, ela é descartada.

```python
from boruta import BorutaPy
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
```

```python
ansur_df = pd.read_csv("../dados/ansur_df.csv")
```

```python
y=ansur_df["Gender"]
y.iloc[:2000]="Female"
X=ansur_df.select_dtypes(include=["int"])
```

```python
boruta = BorutaPy(RandomForestClassifier(), n_estimators='auto', max_iter=15, random_state=7)
```

```python
X.iloc[:, :]
```

```python
boruta.fit(X.values, y)
```

```python
X_new=X.loc[:, boruta.support_]
```
