# Algumas vantagens de redução de dimensionalidade são:
1. Os dados são mais fáceis de trabalhar
2. Requer menos espaço em memória
3. Requer menos tempos computacional
4. Menor chance de modelos de machine learning sofrerem overfitting

# Existem dois tipos de redução de dimensionalidade: Seleção de variáveis e Extração de variáveis.


## Na seleção de variáveis estamos interessados em descartar variáveis irrelevantes para o problemas, enquanto que na extração de variáveis queremos criar novas variáveis a partir das variáveis originais onde o número de variáveis novas é bem menor.
